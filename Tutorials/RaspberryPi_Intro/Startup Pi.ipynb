##Raspian and remote access
Setup raspian.
If it fails you can get back to it by running "raspi-config"  in a terminal window
List of some important commands:
<dl>
<dt> raspi-config <dd> application to help configure the raspian operating system
<dt> startx <dd> script to start up the xwindowing system
<dt> epiphany-browser <dd> Browser for xwindows on the raspberry pi
<dt> wpa-supplicant <dd> application to manage wireless connections
<dt> ifconfig  <dd> application to manage networking on the pi
</dl>

###Networking
#### Ethernet
<p>Attach to an ethernet cable to the ethernet port on your pi.  You may have to reboot for the OS to find the active port. But normally ethernet will connect and assign a ipv4 address using dhcp.  Use ifconfig to find the ip address and write it down for ssh access.  It will be the Inet addr: for the eth0 network entry.
pi-ipaddr=???
Once you start x using startx you will see a few icons in the menu bar on the top right hand of the window. The left most icon at the top of the right the screen gives you access to setup wifi.  This the entry point for the /sbin/wpa_supplicant process that is the application used to setup you network access.
</p>


### Connecting
Make sure ssh is running on the pi
<code>
On the PI run the configuration program:
sudo raspbi-config
select:  8 Advanced Options
Select:  A4 SSH    
It should then say something like: Starting OpenBSD Secure Shell Server: sshd

You can check that sshd is running using:
ps aux | grep sshd

The response should show an application /usr/sbin/sshd is running

On your local computer access the pi using:

ssh pi@xxx.xxx.xxx.xxx
where xxx.xxx.xxx.xxx is the pi-ipaddr you took note earlier.
</code>
You should have changed or should change your password for the pi user so others cannot easily login into your pi.


Now you gave a shell window to run software on the pi on you local machine.
If you do not have ssh on your windows box you can use: putty or cygwin to provide an ssh client.
